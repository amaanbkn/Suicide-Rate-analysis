```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.lineplot(x='year', y='Suicide Rate % change since 2010', data=data, hue='sex', marker='o')
plt.title('Suicide Rate % Change Since 2010 (by Sex)')
plt.xlabel('Year')
plt.ylabel('Suicide Rate % Change')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import pandas as pd
sns.pairplot(data, hue='sex')
plt.show()
```

```python
#Social Media User Change by Year (Twitter vs. Facebook)
data_melted = data.melt(id_vars=['year'], 
                    value_vars=['Twitter user count % change since 2010', 'Facebook user count % change since 2010'],
                    var_name='Social Media', value_name='% Change')
plt.figure(figsize=(10, 6))
sns.barplot(x='year', y='% Change', hue='Social Media', data=data_melted)
plt.title('Social Media User Change by Year (Twitter vs. Facebook)')
plt.xlabel('Year')
plt.ylabel('% Change in User Count')
plt.legend(title='Social Media', loc='upper left')
plt.show()
```
